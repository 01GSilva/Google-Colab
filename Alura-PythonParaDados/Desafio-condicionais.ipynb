1) Escreva um programa que peça à pessoa usuária para fornecer dois números e exibir o número maior.

```python
n1 = float(input('Digite o primeiro numeros: '))
n2 = float(input('Digite o segundo numeros: '))

if n1 > n2:
    print(f'{n1} é o maior valor')
else:
    print(f'{n2} é o maior valor')
```

2) Escreva um programa que solicite o percentual de crescimento de produção de uma empresa e informe se houve um crescimento (porcentagem positiva) ou decrescimento (porcentagem negativa).

```python
percent = float(input('Digite o percentual de crescimento: '))

if percent > 0:
    print('A empresa cresceu.')
elif percent == 0:
    print('Não houve crescimento.')
else:
    print('A empresa diminuiu')
```

3) Escreva um programa que determine se uma letra fornecida pela pessoa usuária é uma vogal ou consoante.

```python
letra = input('Digite alguma letra: ')
vogais = ['a','e','i','o','u']

if letra in vogais:
    print(f'A letra {letra} é uma vogal')
else:
    print(f'A letra {letra} é uma consoante')
```

4) Escreva um programa que leia valores médios de preços de um modelo de carro por 3 anos consecutivos e exiba o valor mais alto e mais baixo entre esses três anos.


```python
v1 = float(input('Digite o valor do carro no primeiro ano: '))
v2 = float(input('Digite o valor do carro no segundo ano: '))
v3 = float(input('Digite o valor do carro no terceiro ano: '))

if v1 > v2 and v1 > v3:
    print(f'O valor mais alto foi {v1}')
elif v2 > v1 and v2 > v3:
    print(f'O valor mais alto foi {v2}')
else:
    print(f'O valor mais alto foi {v3}')

if v1 < v2 and v1 < v3:
    print(f'O valor mais baixo foi {v1}')
elif v2 < v1 and v2 < v3:
    print(f'O valor mais baixo foi {v2}')
else:
    print(f'O valor mais baixo foi {v3}')
```

10) Um programa deve ser escrito para ler dois números e, em seguida, perguntar à pessoa usuária qual operação ele deseja realizar. O resultado da operação deve incluir informações sobre o número - se é par ou ímpar, positivo ou negativo e inteiro ou decimal.

```python
n1 = float(input('Digite o primeiro valor: '))
operacao = input('Qual operação deseja fazer ( + , - , / , *) ?: ')
n2 = float(input('Digite o segundo valor: '))

if operacao == '+':
    res = n1 + n2
    print(f'A soma de {n1} + {n2} é {res}')

elif operacao == '-':
    res = n1 - n2
    print(f'A subtração de {n1} - {n2} é {res}')

elif operacao == '/':
    res = n1 / n2
    print(f'A divisão de {n1} / {n2} é {res}')

elif operacao == '*':
    res = n1 * n2
    print(f'A multiplicação de {n1} * {n2} é {res}')

else:
    print('OPERCAÇÃO INVALIDA')


if res % 2 == 0:
    par_impar = 'par'
else:
    par_impar = 'impar'

if res > 0:
    sinal = 'positivo'
else:
    sinal = 'negativo'

if res - (res // 1) == 0:
    int_dec = 'inteiro'
else:
    int_dec = 'decimal'

print(f'{res} é {par_impar}, {sinal} e {int_dec}')
```
