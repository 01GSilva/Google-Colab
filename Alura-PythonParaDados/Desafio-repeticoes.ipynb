1) Escreva um programa que peça dois números inteiros e imprima todos os números inteiros entre eles.

```python
n1 = int(input('Digite o primeiro numero inteiro: '))
n2 = int(input('Digite o segundo numero inteiro' ))

if n2 < n1:
    aux = n1
    n1 = n2
    n2 = aux
    for i in range(n1+1,n2):
        print(i)

elif n2 == n1:
    print('Valores iguais')
else:
    for i in range(n1+1,n2):
        print(i)
```

2) Escreva um programa para calcular quantos dias levará para a colônia de uma bactéria A ultrapassar ou igualar a colônia de uma bactéria B, com base nas taxas de crescimento de 3% e 1,5% respectivamente. Considere que a colônia A inicia com 4 elementos e a B com 10.



```python
taxa_A = 3/100
taxa_B = 1.5/100

pop_A = 4
pop_B = 10

dias = 1
while pop_A < pop_B:
    pop_A = pop_A + (pop_A*taxa_A)
    pop_B = pop_B + (pop_B*taxa_B)
    dias += 1

print(f'Após {dias} dias, a População A tem {pop_A} individuos e a População B tem {pop_B} individuos ')    
```

3) Para tratar uma quantidade de 15 dados de avaliações de pessoas usuárias de um serviço da empresa, precisamos verificar se as notas são válidas. Então, escreva um programa que vai receber a nota de 0 a 5 de todos os dados e verificar se é um valor válido. Caso seja inserido uma nota acima de 5 ou abaixo de 0, repita até que a pessoa usuária insira um valor válido.

```python
contador = 1
while contador <= 15:
    nota = int(input(f'Digite a {contador}ª nota de avaliação (de 0 a 5): '))
    if nota in range(0,6):
        contador += 1
    else:
        print('Nota invalida!')
```

4) Desenvolva um programa que leia um conjunto indeterminado de temperaturas em Celsius e informe a média delas. A leitura deve ser encerrada ao ser enviado o valor -273°C.

```python
n = 0
soma = 0
temp = 0
while temp != -273:
    temp = float(input('Digite a temperatura em graus Celsius: '))
    soma = soma + temp
    n += 1

media = soma / n
print(f'O programa fez {n} leituras e a média das temperaturas digitadas foi de {media}')
```

5) Escreva um programa que calcule o fatorial de um número inteiro fornecido pela pessoa usuária. Lembrando que o fatorial de um número inteiro é a multiplicação desse número por todos os seus antecessores até o número 1. Por exemplo, o fatorial de 5 é 5 x 4 x 3 x 2 x 1 = 120.



```python
n = int(input('Digite um numero inteiro para calculo do fatorial: '))
res = 1
for i in range(1,n):
    res = res*(i+1)
print(f'O fatorial de {n} é {res}')

```

6) Escreva um programa que gere a tabuada de um número inteiro de 1 a 10, de acordo com a escolha da pessoa usuária. Como exemplo, para o número 2, a tabuada deve ser mostrada no seguinte formato:

7) Os números primos possuem várias aplicações dentro da Ciência de Dados em criptografia e segurança, por exemplo. Um número primo é aquele que é divisível apenas por um e por ele mesmo. Assim, faça um programa que peça um número inteiro e determine se ele é ou não um número primo.



```python
n = int(input('Digite um numero inteiro: '))

for i in range(1, n):
    if (n%i) == 0:
        if (i != 1) and (i != n):
            primo = False
        else:
            primo = True

if primo == True:
    print(f'O número {n} é primo')
else:
    print(f'O número {n} não é primo')
```

8) Vamos entender a distribuição de idades de pensionistas de uma empresa de previdência. Escreva um programa que leia as idades de uma quantidade não informada de clientes e mostre a distribuição em intervalos de [0-25], [26-50], [51-75] e [76-100]. Encerre a entrada de dados com um número negativo.



```python
idade = 0
intervalo_1 = intervalo_2 = intervalo_3 = intervalo_4 = 0

while idade >= 0 :
    idade = int(input('Digite a idade do cliente: '))
    if idade <= 25 and idade >= 0:
        intervalo_1 += 1
    elif idade > 25 and idade <= 50:
        intervalo_2 += 1
    elif idade > 50 and idade <= 75:
        intervalo_3 += 1
    elif idade > 75 and idade <= 100:
        intervalo_4 += 1

print(f'[0-25]: {intervalo_1} / [26-50]: {intervalo_2} / [51-75]: {intervalo_3} / [76-100]: {intervalo_4}')
```

9) Em uma eleição para gerência em uma empresa com 20 pessoas colaboradoras, existem quatro candidatos(as). Escreva um programa que calcule o(a) vencedor(a) da eleição. A votação ocorreu da seguinte maneira:

1. Cada colaborador(a) votou em uma das quatro pessoas candidatas (que representamos pelos números 1, 2, 3 e 4).
2. Também foram contabilizados os votos nulos (representados pelo número 5) e os votos em branco (representados pelo número 6).

```python
cand_1 = cand_2 = cand_3 = cand_4 = nulos = branco = 0
contador = 1
total = 10

while contador <= total:
    voto = int(input(f'Funcionario {contador} digite seu voto: '))
    if voto <= 6:

        if voto == 1:
            cand_1 += 1

        if voto == 2:
            cand_2 += 1

        if voto == 3:
            cand_3 += 1

        if voto == 4:
            cand_4 += 1

        if voto == 5:
            nulos += 1

        if voto == 6:
            branco += 1

        contador += 1
    else:
        print('Voto Invalido!')
        contador = contador

print('RESULTADOS')
print(f'TOTAL DE VOTOS: {total}')
print(f'CANDIDATO 1: {cand_1} VOTOS')
print(f'CANDIDATO 2: {cand_2} VOTOS')
print(f'CANDIDATO 3: {cand_3} VOTOS')
print(f'CANDIDATO 4: {cand_4} VOTOS')
print(f'NULOS: {nulos} VOTOS')
print(f'BRANCOS: {branco} VOTOS')


perc_branco = float((branco * 100) / total)
perc_nulos = float((nulos * 100) / total)

print(f'PORCENTAGEM DE BRANCOS: {perc_branco}%')
print(f'PORCENTAGEM DE NULOS: {perc_nulos}%')
    
```
